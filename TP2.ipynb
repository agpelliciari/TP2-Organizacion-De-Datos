## Cantidad de productos vendidos por mes desde 01/01/2013

```python
ventas_totales_por_mes = sales.groupby('date_block_num')['item_cnt_day'].sum().to_frame().reset_index()
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=ventas_totales_por_mes, x="date_block_num", y="item_cnt_day")

plt.title('Variacion de la cantidad de productos vendidos por mes')

plt.xlabel('Meses despues del 01/01/2013')
plt.xticks(ventas_totales_por_mes['date_block_num'])

plt.ylabel('Cantidad de productos vendidos')
plt.ylim(0)

plt.tight_layout()
plt.show()
```

```python
# Límites de los intervalos: <30, 30-90, 90-180, >180
bins = [0, 12, 24, 34]

# Definir las etiquetas para los intervalos
labels = ['2013', '2014', '2015']

# Agrupar los datos por la columna 'purchase_lead' en los intervalos especificados
ventas_totales_por_mes['anio'] = pd.cut(ventas_totales_por_mes['date_block_num'], bins=bins, labels=labels, right=False)
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=ventas_totales_por_mes, x="date_block_num", y="item_cnt_day", hue="anio")

plt.title('Variacion de la cantidad de productos vendidos por mes')

plt.xlabel('Meses despues del 01/01/2013')
plt.xticks(ventas_totales_por_mes['date_block_num'])

plt.ylabel('Cantidad de productos vendidos')
plt.ylim(0)

plt.tight_layout()
plt.show()
```

```python
def modificando(numero):
    if (numero < 12):
        return numero + 1
    elif (numero < 24):
        return (numero - 12) + 1
    else: 
        return (numero - 24) + 1  
```

```python
ventas_totales_por_mes['date_block_num'] = ventas_totales_por_mes['date_block_num'].apply(modificando)
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=ventas_totales_por_mes, x="date_block_num", y="item_cnt_day", hue="anio")

plt.title('Variacion de la cantidad de productos vendidos por mes')

plt.xlabel('Mes')
plt.xticks(ventas_totales_por_mes['date_block_num'])

plt.ylabel('Cantidad de productos vendidos')
plt.ylim(0)

plt.tight_layout()
plt.show()
```

# Baseline (Regresion Lineal)

### Separar la variable a predecir -> **item_cnt_day** (La cantidad de ese item que fue vendida x dia)

```python
sales = sales.groupby(['date_block_num', 'shop_id', 'item_id'])['item_cnt_day'].sum().to_frame().reset_index()
```

```python
sales
```

```python
x = sales.drop('item_cnt_day', axis=1)
```

```python
y = sales['item_cnt_day']
```

```python
x
```

```python
y
```

### Division de los nuevos sets -> **train, test y validation**

```python
train_set_size = 0.8
val_set_size = 0.2
set_size = len(x) # tamaño del set de datos
train_index_end = int(train_set_size * set_size) # indice donde finaliza el set de entrenamiento
val_index_end = train_index_end + int(val_set_size * set_size) # indice donde finaliza el set de validacion y testeo

(0, train_index_end, val_index_end, set_size)
```

```python
def dividir_set_de_datos(dataset, train_index_end):
    return (dataset.iloc[:train_index_end], dataset.iloc[train_index_end:] )
```

```python
x_train_set, x_validation_set = dividir_set_de_datos(x, train_index_end)
```

```python
y_train_set, y_validation_set = dividir_set_de_datos(y, train_index_end)
```

- **set de entrenamiento**: desde 0 hasta el indice donde finaliza el set de entrenamiento (80% del set)
- **set de validacion**: desde el indice donde finaliza el set de entrenamiento hasta el indice donde finaliza el set de validacion (10% del set)
- **set de testeo**: desde el indice donde finaliza el set de validacion hasta el final del set de datos (10% del set)

### Preprocesamiento del set -> **Feature Engineering**

- fecha

```python
import numpy as np
```

```python
x_train_set['date'] = pd.to_datetime(x_train_set['date'], format='%d.%m.%Y', dayfirst=True)
```

```python
x_validation_set['date'] = pd.to_datetime(x_validation_set['date'], format='%d.%m.%Y', dayfirst=True)
```

```python
x_train_set['date:year'] = x_train_set['date'].dt.year
x_train_set['date:month'] = x_train_set['date'].dt.month
x_train_set['date:day'] = x_train_set['date'].dt.day
x_train_set['date:day_of_week'] = x_train_set['date'].dt.day_of_week
x_train_set['date:day_of_year'] = x_train_set['date'].dt.day_of_year
```

```python
x_validation_set['date:year'] = x_validation_set['date'].dt.year
x_validation_set['date:month'] = x_validation_set['date'].dt.month
x_validation_set['date:day'] = x_validation_set['date'].dt.day
x_validation_set['date:day_of_week'] = x_validation_set['date'].dt.day_of_week
x_validation_set['date:day_of_year'] = x_validation_set['date'].dt.day_of_year
```

```python
x_train_set['date:is_weekend'] = np.where(x_train_set['date:day_of_week'].isin([5,6]), 1,0)
```

```python
x_validation_set['date:is_weekend'] = np.where(x_validation_set['date:day_of_week'].isin([5,6]), 1,0)
```

```python
x_train_set = x_train_set.drop('date', axis=1)
```

```python
x_validation_set = x_validation_set.drop('date', axis=1)
```

- precio promedio por item

```python
x_train_set['mean_price_per_item'] = x_train_set.groupby('item_id')['item_price'].transform('mean')
```

```python
x_validation_set['mean_price_per_item'] = x_validation_set.groupby('item_id')['item_price'].transform('mean')
```
